```python
#阴阳师游戏壁纸获取

#导包
import os
import csv
import requests
from lxml import etree
```

```python
#User-Agent换成自己的就行
url = 'https://yys.163.com/media/picture.html'
headers = {
    'User-Agent':'换上自己的',
    'Referer':'https://yys.163.com/media/picture.html'
}
response = requests.get(url,headers=headers)
content = response.content.decode('utf8')
html = etree.HTML(content)
print(content)
```

```python
#爬取链接和对应的图片尺寸

urls = html.xpath('//div[@class="part-2"]//div[@class="item"]//a/@href')
sizes = html.xpath('//div[@class="part-2"]//div[@class="item"]//a/text()')
print(len(urls),'\n',urls)
print(len(sizes),'\n',sizes)
#这里只爬取“手机壁纸”、“横板”、“竖版”的内容，不爬取封面
```

```python
#构建链接与尺寸之间的字典

onmyoji = []
for url, size in zip(urls, sizes):
    eg = {
        'url':url,
        'size':size
    }
    onmyoji.append(eg)
onmyoji
```

```python
#将链接和对应的尺寸保存到本地为csv文件

#保存路径
save_path = r'F:\ProgramData\Crawling_class\onmyoji\documents\onmyoji_data.csv'
# 保存为CSV文件
with open(save_path, mode='w', newline='', encoding='utf-8-sig') as f:
    writer = csv.DictWriter(f, fieldnames=['url', 'size'])
    writer.writeheader()
    writer.writerows(onmyoji)

print(f'保存成功，文件路径：{save_path}')
```

```python
#下载图片到本地文件夹

#按照图片尺寸进行分类

#首先确定图片有哪些尺寸，使用集合进行去重

base_dir = r'F:\ProgramData\Crawling_class\onmyoji\images'  #保存路径
#统一sizes格式为str

size_class = set(str(item['size']) for item in onmyoji)

#为每一个尺寸创建文件夹
for size in size_class:
    folder_path = os.path.join(base_dir, size)
    os.makedirs(folder_path, exist_ok=True)

#下载并保存图片
for i, item in enumerate(onmyoji, 1):
    url = item['url']
    size = item['size']
    try:
        response = requests.get(url, stream=True, timeout=10)
        response.raise_for_status()

        ext = os.path.splitext(url)[-1] or '.png'
        filename = f'image_{i}{ext}'
        file_path = os.path.join(base_dir, size, filename)

        with open(file_path, 'wb') as f:
            for chunk in response.iter_content(1024):
                f.write(chunk)

        print(f"已保存到 {file_path}")
    except Exception as e:
        print(f"下载失败 [{url}]: {e}")
```
